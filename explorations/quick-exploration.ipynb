```python
from pathlib import Path

import numpy as np
import pandas as pd
from datetime import datetime 

%matplotlib inline
```

```python
data_dir = Path("../data/")
df = pd.read_csv(data_dir/"Business_Licenses.csv")
```

```python
df["year"] = pd.to_datetime(df["DATE ISSUED"]).dt.year
df
```

```python
df.columns
```

```python
df.drop(columns=['LATITUDE', 'LONGITUDE', 'LOCATION', 'ADDRESS', 'CITY', 'STATE', 'ZIP CODE',
       'WARD', 'PRECINCT', 'WARD PRECINCT', 'POLICE DISTRICT'], inplace=True)
```

```python
df
```

```python
df.head(5000).groupby(["ACCOUNT NUMBER", "LEGAL NAME"])["year"].apply(list).apply(sorted)
```

```python
years_by_license_number = df.head(10000).groupby(["LICENSE NUMBER"])["year"].apply(list).apply(sorted).to_frame()
years_by_license_number[years_by_license_number["year"].apply(len) > 1]
```

```python
renewals = df[
    (df["APPLICATION TYPE"] == "RENEW") & # subsequent renewal
    (df["LICENSE STATUS"]   == "AAI")     # license granted
].head(10000)

renewals_by_year = (
    renewals.groupby(["LICENSE NUMBER"])["year"]
    .apply(list)
    .apply(sorted)
    .to_frame())

multiple_renewals = renewals_by_year[renewals_by_year["year"].apply(len) > 1]
multiple_renewals = multiple_renewals.rename({"year": "renewal years"})

renewed_records = df.join(
    multiple_renewals["year"],
    on="LICENSE NUMBER",
    how="right", 
    rsuffix="_r")

renewed_records
```
